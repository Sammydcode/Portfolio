Price-to-Earnings (P/E) Ratio

- Formula: P/E Ratio = Current Share Price / Earnings per Share
- Explanation: This formula calculates the ratio of a company's current share price to its earnings per share.
- Relationship to the project: Companies with a low P/E ratio may be undervalued, while those with a high P/E ratio may be overvalued.
- Selection criteria: Select companies with a reasonable P/E ratio (e.g., between 10 and 30).


Industry Trends (VISUALIZATION)

### Profitability Margin

```python
#Checking the Data type of the df
df.dtypes
```

```python
# Company Performance Analysis

#Calculating the profitability margin = (profit/sales)*100

df["Profitability Margin"] = (df["Profit($)"]/df["Sales($)"])*100
df
```

```python
#picking the full rows of all the companies with a positive profitability margin 

sorted_df = df.loc[df["Profitability Margin"] > 0]
sorted_df
```

**FROM THE PROFIT MARGIN (PROFITABILITY ANALYSIS), COMPANY WITH A NON NEGATIVE PROFITABILITY MARGIN HAS BEEN SELEDCTED AND SAVED IN THE VARIABLE CALLEED SORTED DF**

### Sales per Country

Sales per Country

- Formula: Sales per Country = Total Sales / Number of Countries
- Explanation: This formula calculates the average sales per country for each company.
- Relationship to the project: Companies with high sales per country may have a strong market presence in specific regions.
- Selection criteria: Companies with Sales per Country > $100 million


```python
#Calculating and adding the Sales per Country column to the data frame

sorted_df.loc[:, "Sales per Country"] = sorted_df["Sales($)"]/(sorted_df["Country"]).nunique()
sorted_df
```

```python
#sorting out  companies with a sales per country greater than 100 Million

new_sorted_df = sorted_df.loc[sorted_df["Sales per Country"]> 100000000]
new_sorted_df
```

### FROM THE SALES PER COUNTRY, COMPANY WITH A SALES PER COUNTRY GREATER THAN $100 MILLION  HAS BEEN SELECCTED AND SAVED IN THE VARIABLE CALLED NEW SORTED DF

### Market Capitalization

Market Capitalization

- Formula: Market Capitalization = Total Number of Shares x Current Share Price; Market Capitalization is also a substitute for Market value
- Explanation: This formula calculates the total value of a company's outstanding shares.
- Relationship to the project: Companies with a high market capitalization/ value may have a strong financial foundation.
- Selection criteria: Selecting Companies with a market value > $1 billion


```python
updated_df = new_sorted_df.loc[new_sorted_df["Market Value($)"] > 1000000000]
updated_df
```

### FROM THE DF, COMPANY WITH A MARKET VALUE GREATER THAN $1 BILLION HAS BEEN SELECCTED AND SAVED IN THE VARIABLE CALLED: UPDATED DF


### Price to Sales Ratio

Price-to-Sales (P/S) Ratio:

- Explaination: This ratio helps evaluate a company's value relative to its sales.
- Formular: Calculate this by dividing the market value by the company's sales.
- Relationship to the Project:
Companies with a low P/S ratio may indeed be undervalued, as their market value is relatively low compared to their sales. Conversely, companies with a high P/S ratio may be overvalued, as their market value is relatively high compared to their sales.
- Selection Criteria:
Selecting companies with a reasonable P/S ratio, such as between 1 and 3, or 1 and 5 depending on the industry, can help identify those that are fairly valued. However, the specific range of 10 to 30 might be too narrow or high, depending on the industry and market conditions.

```python
#Adding a new column "Price to sales ratio"
updated_df.loc[:, "Price to Sales Ratio"] = updated_df["Market Value($)"]/updated_df["Sales($)"]
updated_df
```

```python
#sorting out companies with price to sale ration between 1 and 5 and saving it in a variable called "Updated df 1"

updated_df_1 = updated_df.loc[(updated_df["Price to Sales Ratio"] <= 5) & (updated_df["Price to Sales Ratio"] >= 1) ]
updated_df_1
```

### FROM THE DF, COMPANY WITH A PRICE TO SALE RATIO BETWEEN 1 AND 5 HAS BEEN SELECCTED AND SAVED IN THE VARIABLE CALLED: UPDATED DF 1

4. Cluster dominance: The points in the cluster appear more dense and numerous than the outliers.

**Insights and Potential Implications**

1. Efficient companies: The initial outliers might represent companies with high market values despite low profitability margins, potentially indicating inefficient operations or other factors.
2. Healthy growth: The positive correlation cluster could suggest that companies with higher profitability margins tend to have higher market values, indicating healthy growth and a strong relationship between these variables.
3. Overvalued companies: The second set of outliers might represent companies with high profitability margins but low market values, potentially indicating undervaluation or other market inefficiencies.
4. Market dynamics: The dominance of the cluster points might suggest that the market tends to reward companies with balanced profitability and market value, while outliers might be subject to various market forces and inefficiencies.


**Observation**

A review of the scatter plot between profitability margin and market value reveals an interesting pattern. Two distinct groups of outliers emerge:

- **_High Profitability, Low Market Value_**
